```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('preprocessed2.csv')
```

```python
df = df.rename(columns = {'Number of Bedrooms': 'No_of_Bedrooms'}, inplace = False)
df = df.rename(columns = {'Property Type': 'Property_Type'}, inplace = False)
```

```python
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
Number_of_Bedrooms	=pd.get_dummies(df['No_of_Bedrooms'])
```

```python
Property_type=pd.get_dummies(df['Property_Type'])
```

```python
Location_new=pd.get_dummies(df['Location'])
```

```python
df=pd.concat([df,Location_new,Number_of_Bedrooms,Property_type],axis=1)
```

```python
df.drop(['STATISTIC','Statistic','Year','TLIST(A1)','No_of_Bedrooms','C02969V03591','Property_Type','C02970V03592','UNIT','Location','C03004V03625'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Train data
X=df.drop('VALUE',axis=1)
Y=df['VALUE']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Fitting Decision Tree Regression to the dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print("R-Sqaured: %.3f"
      % regressor.score(X_test,y_test))
```

```python
print("Mean squared error: %.3f"
      % mean_squared_error(y_test, y_pred))  # Test MSE

print("Mean absolute error: %.3f"
      % mean_absolute_error(y_test, y_pred))
```
